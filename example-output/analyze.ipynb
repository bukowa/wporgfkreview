```python
import json

data = json.load(open('plugins.json'))
reviews = [
    {**y, 'plugin': x['Url']}  # Add 'plugin' key to each review
    for x in data              # Iterate over each plugin
    for y in x['Reviews']      # Iterate over each review in 'Reviews'
]
len(reviews)
```

```python
import pandas as pd
df = pd.DataFrame(reviews)

df.head()
```

```python
from datetime import datetime

df['AuthorCreatedDate'] = pd.to_datetime(df['AuthorCreated'],errors='coerce')
df['ReviewCreatedDate'] = pd.to_datetime(df['ReviewCreated'],errors='coerce')
df['Diff'] = df['ReviewCreatedDate'] - df['AuthorCreatedDate']
df_final = df[['AuthorCreatedDate', 'ReviewCreatedDate', 'Diff', 'AuthorUrl', 'Url', 'plugin']].sort_values('Diff')
df_final
```

```python
df_final.to_csv('diff.csv')
```
